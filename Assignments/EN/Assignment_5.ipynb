#### 1. Count words in a text

a. Output a list of words in the file along with their frequency counts (ignoring case).   
a. Count how many unique words there are (ignoring case).    
c. Check how common are all different sequences of vowels (e.g. the sequences "ieu" or just "e" in "lieutenant")?

```python
# a)
# adittionally removed suffix 's after words before the count for company's was 35 no for company it is 272
lower_strip_tokens = [token.strip(string.whitespace).strip(string.punctuation).removesuffix("\'s") for token in text.lower().split()]

# removed empty list-items
cleaned_lower_tokens = []
i0 = 0
while i0 < len(lower_strip_tokens):
    if lower_strip_tokens[i0] == "":
        pass
    else:
        cleaned_lower_tokens.append(lower_strip_tokens[i0])
    i0 += 1
    
counted_words={}
for word in cleaned_lower_tokens:
    if word in counted_words:
        counted_words[word] += 1
    else:
        counted_words[word] = 1
print(counted_words)
```

```python
# b)
print(len(set(cleaned_lower_tokens)))
```

```python
# c)
vowels = 'AEIOUaeiou'
i = 0
sequences = {}
while i < len(text):
    seq = ''
    while i < len(text) and text[i] in vowels:
        seq += text[i]
    # ignoring case:
    #       seq += text[i].lower()
        i += 1 
    if seq in sequences:
        sequences[seq] += 1
    elif seq == '':
        pass
    else:
        sequences[seq] = 1
    i += 1

#print(sequences)
print(sorted(sequences.items(), key=lambda item: item[1], reverse=True))
```

```python
# c)
# including empty lines
print(f"{len(text.splitlines())} lines (including empty lines)")
# cleaned_lines list from exercise 2.a)
print(f"{len(cleaned_lines)} printed lines")

strip_tokens = [token.strip(string.whitespace).strip(string.punctuation) for token in text.split()]
# remove empty tokens from tokens list
cleaned_tokens = []
i3 = 0
while i3 < len(strip_tokens):
    if strip_tokens[i3] == "":
        pass
    else:
        cleaned_tokens.append(strip_tokens[i3])
    i3 += 1
# maybe numbers should be removed too
print(f"{len(cleaned_tokens)} printed words (including numerical values)")

# number of characters including special characters but excluding whitespace
print(f'{len("".join(text.replace(" ", "").split()))} characters including special characters but excluding whitespace')

```

```python
# d)
words = [token.strip(string.whitespace).strip(string.punctuation) for token in text.split()]

all_upper = []
for word in words:
    if word.isupper():
        all_upper.append(word)
        
print(f"{len(all_upper)} all uppercase words")
print(all_upper)
    

```

```python
# e)
# only included words with 4 alphabetic values, excluding "it's", "50kg", etc.
# words list from exercise 3.d)
four_char = []
for word in words:    
    if word.isalpha() and len(word) == 4:
        four_char.append(word)
        

print(f"{len(four_char)} four character words")
print(four_char)
```

```python
# f)
# only included tokens consisting of more than one alphabetic characters 
no_vowels = []
for word in words:
    if not re.search(r"[AEIOUaeiou]",word) and word.isalpha() and len(word)>1:
        no_vowels.append(word)

print(f"{len(set(no_vowels))} different words with no vowels in them")
print(set(no_vowels))
```

```python
# g)
# solution is only an approximation. still many exceptions to be caught
one_syl = []
for word in cleaned_tokens:
    count = 0
    vowels = "aeiouyAEIOUY"
    if word[0] in vowels:
        count += 1
    for i in range(1, len(word)):
        if word[i] in vowels and word[i - 1] not in vowels:
            count += 1
        if word[-3:] == "ing" and word[-4] in vowels:
            count += 1
    if word.endswith("e") and not word.endswith("le") and not word.endswith("ee") and not word.endswith("ie"):
        count -= 1
    if count == 1:
        one_syl.append(word)
print(one_syl)
```

#### 4. Compute ngrams  

a. Find the 10 most common bigrams  
b. Find the 10 most common trigrams  

```python
# a)
# resulting dictionary counts word-bigrams
bigrams={}
i=0
while i < len(cleaned_tokens)-1:
    if (cleaned_tokens[i] + " " + cleaned_tokens[i+1]) in bigrams:
        bigrams[cleaned_tokens[i] + " " + cleaned_tokens[i+1]] += 1
    else:
        bigrams[cleaned_tokens[i] + " " + cleaned_tokens[i+1]] = 1
    i += 1
print(sorted(bigrams.items(), key=lambda item: item[1], reverse=True)[:10])
```

```python
# b)
# resulting dictionary counts word-trigrams
trigrams={}
i=0
while i < len(cleaned_tokens)-2:
    if (cleaned_tokens[i] + " " + cleaned_tokens[i+1] + " " + cleaned_tokens[i+2]) in trigrams:
        trigrams[cleaned_tokens[i] + " " + cleaned_tokens[i+1] + " " + cleaned_tokens[i+2]] += 1
    else:
        trigrams[cleaned_tokens[i] + " " + cleaned_tokens[i+1] + " " + cleaned_tokens[i+2]] = 1
    i += 1
print(sorted(trigrams.items(), key=lambda item: item[1], reverse=True)[:10])
```

```python
with open("../../Data/txt/secret_ec.txt", "r") as f:
    text_extra = f.read()

tokens = text_extra.split()
    
counted_words={}
for word in tokens:
    if "e" in word and word in counted_words:
        counted_words[word] += 1
    else:
        counted_words[word] = 1
print(sorted(counted_words.items(), key=lambda item: item[1], reverse=True)[:2])
```

##### Extra Credit Exercise 2
• Find the 2 most common bigrams in secret_ec.txt where the second word in the bigram ends with a consonant.  
• Your answer will correspond to the next four words of the secret message.  

```python
vowels="aeiou"
bigrams={}
i=0
while i < len(tokens)-1:
    if (tokens[i] + " " + tokens[i+1]) in bigrams and not tokens[i+1][-1] in vowels:
        bigrams[tokens[i] + " " + tokens[i+1]] += 1
    else:
        bigrams[tokens[i] + " " + tokens[i+1]] = 1
    i += 1
print(sorted(bigrams.items(), key=lambda item: item[1], reverse=True)[:2])
```
